```python
import numpy as np
import matplotlib.pyplot as plt

f= lambda x:(np.tan(x))**0.5

def derivada_exacta(x):
    return 1/(2*((np.tan(x))**0.5)*((np.cos(x))**2))
def derivada_progresiva (f,x,h=0.01):
    return ((f(x+h)-f(x))/h)
def derivada_central (f,x,h=0.01):
    return ((f(x+h)-f(x-h))/(2*h)) #Calculado Analiticamente(punto e)

def error_nodal(y1,y2):
    return np.abs(y1-y2)
intervalo=np.linspace(0.1,1.1,100)

lista_derivadas_progresivas=[]
lista_derivadas_centrales=[]
lista_derivadas_exactas=[]

lista_errores_progresivas=[]
lista_errores_centrales=[]

for i in intervalo:
    d_progresiva=derivada_progresiva(f,i)
    d_central=derivada_central(f,i)
    d_exacta=derivada_exacta(i)
    lista_derivadas_progresivas.append(d_progresiva)
    lista_derivadas_centrales.append(d_central)
    lista_derivadas_exactas.append(d_exacta)
    error_progresiva=error_nodal(d_progresiva,d_exacta)
    error_central=error_nodal(d_central,d_exacta)
    lista_errores_progresivas.append(error_progresiva)
    lista_errores_centrales.append(error_central)
    
    
fig=plt.figure(figsize=(20,10))
fig2=plt.figure(figsize=(10,10))
ax=fig.add_subplot(1,1,1)
ax1=fig2.add_subplot(1,1,1)
ax.scatter(intervalo,lista_derivadas_progresivas)
ax.scatter(intervalo,lista_derivadas_centrales)
ax.scatter(intervalo,lista_derivadas_exactas)
ax1.scatter(intervalo,lista_errores_progresivas)
ax1.scatter(intervalo,lista_errores_centrales)
```
