```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

```python
def CreateSphere( N, R ):
    
    X = np.zeros(N)
    Y = np.zeros_like(X)
    Z = np.zeros_like(X)
    
    for i in tqdm(range(N)):
        
        u = np.random.rand()
        r = R*u**(1/2)
        
        phi = np.random.uniform(0,2*np.pi)
        theta=np.random.uniform(0,np.pi)
        
        X[i] = r*np.sin(theta)*np.cos(phi)
        Y[i] = r*np.sin(theta)*np.sin(phi)
        Z[i] = r*np.cos(theta)
    
    
    return X,Y,Z

X,Y,Z = CreateSphere(10000, 1)

```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, projection = '3d')

#configurar límites de los ejes
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(0, 2)

ax.view_init(10, 60)

ax.scatter(X,Y,Z,color='b')
```

```python
def f(R):
    return np.e**R
```

```python
def GetPoints(R):
    
    u = np.random.rand()
    r = R*u**(1/2)    
    
    return r
```

```python
N = int(1e6)
Sample = np.zeros(N)
for i in range(N):
    r= GetPoints(1)
    Sample[i] = f(r)
```

```python
Integral = 2*np.pi * np.average( Sample )
```
