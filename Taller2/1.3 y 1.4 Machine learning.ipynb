```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import sympy as sy
```

```python
M = [1.0,0.0,-1.0]
m_1 = np.array(M)
h=0.05
x=np.linspace(-10,10,1000)
```

```python
def func(x):
    f_x = 1/(np.sqrt(1+(np.exp(-x**2))))
    return f_x
    
def derivada_central(f,x,M,h):
    sum=0
    for i in M:
        sum -= f(x-i*h)*i
    sum /= (2*h)
    return sum

def segunda_Dcentral(f,x,M,h):
    sum=0
    for i in M:
        miniSum = 0
        for j in M:
            miniSum -= f(x-i*h-j*h)*j
        sum -= i*miniSum
    sum /= (2*h)**2
    return sum
            
```

```python
y = func(x)
dy = derivada_central(func,x,m_1,h)
d2y = segunda_Dcentral(func,x,m_1,h)
```

```python
plt.plot(x,y)
plt.title("y vs x (sin derivar)")
plt.show
```

```python
plt.plot(x,dy)
plt.title("dy vs x (derivada)")
plt.show
```

```python
plt.plot(x,d2y)
plt.title("d2y vs x (derivada)")
plt.show
```
