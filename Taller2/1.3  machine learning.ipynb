```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
M = [1.0,0.0,-1.0]
m_1 = np.array(M)
h=0.05
x=np.linspace(-10,10,1000)
```

```python
def funcion_2(x):
    f_x = 1/(np.sqrt(1+(np.exp(-x**2))))
    return f_x
    
def derivada_central(f,x,M,h):
    sum=0
    for i in M:
        sum -= f(x-i*h)*i
    sum /= (2*h)
    return sum
```

```python
y = funcion_2(x)
dy = derivada_central(funcion_2,x,m_1,h)
```

```python
plt.plot(x,y)
plt.plot(x,dy)
```
