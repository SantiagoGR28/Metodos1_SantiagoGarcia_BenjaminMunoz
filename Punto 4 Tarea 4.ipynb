```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm 
```

```python
N=10**4
x=np.random.uniform(0,1000,N+30)
k=30
X=[]
Y=[]
C=0
suma=0

for j in range(1,k):
    for i in range(1,N):
        suma+=((x[i])*(x[i+j]))
    X.append(j)
    Y.append(suma/N)
    print(Y)

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax1.plot(X, Y)
        
```
