```python
#3. Números factoriales
#3.1
def factorial(numero): 
    if numero < 0: 
        print("No existe el numero factorial de números negativos")

    elif numero == 0: 
        return 1
        
    else: 
        factorial = 1
        while(numero > 1): 
            factorial *= numero 
            numero -= 1
        return factorial 

numero = 5; 

print("El número factorial de",numero,"es", factorial(numero)) 

```

```python
#3.2
def repeticiones(r, n):
    
    div = factorial(n)
    if r < 0: 
        return("No existe el numero factorial de números negativos")
    resta = (n-r)
    divs = factorial (resta)
    
    resultado = div/divs
    return resultado
    
r = 3
n = 6

print(repeticiones(r, n))
```

```python
#3.3
def combinaciones(n, m):
    if n > m:
        resta = (n - m)
        n = factorial(n)
        m = factorial (m)
        resta = factorial(resta)
        resultado = n/(m*(resta))
        return(resultado)
    else:
        return("No es posible calcular este resultado.")
    
print(combinaciones(22,11))
print(combinaciones(21,10))
```

```python
#5. Sucesión de Fibonacci
#5.1
def Fibonacci1(n):
    if(n==0):
        return 0
    elif(n==1):
        return 1
    else:
        return (Fibonacci(n-2)+Fibonacci(n-1))
    
def Fibonacci2(n): 
    lst=[]
    for n in range(0, n+1):
        lst.append(Fibonacci1(n))
    return(lst)
        
n = 20     
print(Fibonacci2(n))
```

```python
#5.2 Gráfica de sucesión de Fibonacci
#n en eje x y f(n) en eje y
import matplotlib.pyplot as plt
import numpy as np
n = 20
x= np.arange(0,n+1,1)
y= Fibonacci2(n)
print(x)
print(y)
plt.plot(x,y)
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('Sucesión de Fibonacci')
plt.show()
```

```python
#numeros primos 1
import numpy as np
def primos():
    lista_numeros=np.arange(1,1000)
    lista_p=[]

    for i in range( len(lista_numeros)):
        contador=0
        numero=lista_numeros[i]
        for j in range(len(lista_numeros)):
            if (numero%lista_numeros[j]==0):
                contador=contador+1
        if contador==2:
            lista_p.append(numero)
    return lista_p
    
#numeros primos 2
lista = primos()
print(lista[0:10])

#numeros primos 3
x= np.arange(0,10,1)
y= lista[0:10]
plt.plot(x,y)
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('Posición de los Números Primos')
plt.show()
```

```python
from math import pi
from math import sin
from math import cos
import numpy as np

ang = []
for f in range(-500, 1000):
    ang.append((f*pi)/180)

a = 0
b = 1
pos =[]
for i in range(0, len(ang)):
    r = a + b*ang[i]
    pos.append(r)


x = []
for a in range(0, len(ang)):
    x.append(pos[a]*cos(ang[a]))
    
y = []
for b in range(0, len(ang)):
    y.append(pos[b]*sin(ang[b]))

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Espiral de Arquímedes')
plt.show()
```
