```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=1
theta=np.linspace(0,2*np.pi,200)
desfase=[0,np.pi/4,np.pi/2]
nsides=5
nsides1=5
nsides2=5
k=1
l=1

print("Figuras desfase=0")
for i in range(0,5):
    nx=i+1
    for j in range(nx-1,5):
        ny=j+1
        x=np.sin(nx*theta)
        y=np.sin(ny*theta+desfase[0])
        if ny>=nx:
            fig = plt.subplot(nsides,nsides,k)
            fig=plt.plot(x,y)
            plt.show()
            k += 1
k=1
print("Figuras desfase=pi/4")
for i in range(0,5):
    nx=i+1
    for j in range(nx-1,5):
        ny=j+1
        x=np.sin(nx*theta)
        y=np.sin(ny*theta+desfase[1])
        if ny>=nx:

            fig = plt.subplot(nsides,nsides,k)
            plt.plot(x,y)
            plt.show()
            k += 1
k=1
print("Figuras desfase=pi/2")
for i in range(0,5):
    nx=i+1
    for j in range(nx-1,5):
        ny=j+1
        x=np.sin(nx*theta)
        y=np.sin(ny*theta+desfase[2])
        if ny>=nx:

            fig = plt.subplot(nsides,nsides,k)
            plt.plot(x,y)
            plt.show()
            k += 1
```
