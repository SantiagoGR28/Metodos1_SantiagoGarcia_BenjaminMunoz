```python
from math import pi
from math import sin
from math import cos
import numpy as np
import matplotlib.pyplot as plt
ang = []
for f in range(-500, 1000):
    ang.append((f*pi)/180)

a = 0
b = 1
pos =[]
for i in range(0, len(ang)):
    r = a + b*ang[i]
    pos.append(r)


x = []
for a in range(0, len(ang)):
    x.append(pos[a]*cos(ang[a]))
    
y = []
for b in range(0, len(ang)):
    y.append(pos[b]*sin(ang[b]))

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Espiral de Arqu√≠medes')
plt.show()
```
