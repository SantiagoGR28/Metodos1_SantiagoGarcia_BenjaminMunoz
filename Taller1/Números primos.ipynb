```python
#numeros primos 1
import numpy as np
import matplotlib.pyplot as plt
def primos():
    lista_numeros=np.arange(1,1000)
    lista_p=[]

    for i in range( len(lista_numeros)):
        contador=0
        numero=lista_numeros[i]
        for j in range(len(lista_numeros)):
            if (numero%lista_numeros[j]==0):
                contador=contador+1
        if contador==2:
            lista_p.append(numero)
    return lista_p
    
#numeros primos 2
lista = primos()
print(lista[0:10])

#numeros primos 3
x= np.arange(0,10,1)
y= lista[0:10]
plt.plot(x,y)
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('Posición de los Números Primos')
plt.show()
```
